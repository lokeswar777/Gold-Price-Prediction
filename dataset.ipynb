## Creating Dataset for past year

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import time

def fetch_gold_price_for_date(date):
    api_key = "goldapi-9z4dsm6vsntr2-io"
    symbol = "XAU"
    curr = "INR"

    # Date format should be YYYY-MM-DD
    url = f"https://www.goldapi.io/api/{symbol}/{curr}/{date}"

    headers = {
        "x-access-token": api_key,
        "Content-Type": "application/json"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        data = response.json()
        
        gold_data = {
            'date': date,
            'price': data.get('price'),
            'open_price': data.get('open_price'),
            'high_price': data.get('high_price'),
            'low_price': data.get('low_price')
        }

        return gold_data

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for {date}: {e}")
        return None

def create_gold_price_dataset():
    # Set today's date and calculate date 365 days ago
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)
    
    dataset = []

    current_date = start_date
    while current_date <= end_date:
        formatted_date = current_date.strftime('%Y-%m-%d')
        gold_data = fetch_gold_price_for_date(formatted_date)

        if gold_data:
            dataset.append(gold_data)
            print(f"Data for {formatted_date} fetched successfully.")

        # Sleep to avoid hitting API rate limits
        time.sleep(1)  # Adjust if needed depending on API limits
        
        current_date += timedelta(days=1)
    
    # Convert to DataFrame and save to CSV
    df = pd.DataFrame(dataset)
    df.to_csv('gold_price_dataset.csv', index=False)
    print("Gold price dataset saved as 'gold_price_dataset.csv'.")

# Run the data collection
create_gold_price_dataset()
```
